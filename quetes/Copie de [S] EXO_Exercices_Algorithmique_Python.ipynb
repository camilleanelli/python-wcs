## Exercice 3 : Vérification si un nombre est pair ou impair
Écrire un programme Python qui vérifie si un nombre donné est pair ou impair.


```python
number = 5
if(number % 2 == 0):
    print(f"{number} est pair")
else:
    print(f"{number} est impair")
```

## Exercice 4 : Calcul de la factorielle d'un nombre
Écrire un programme Python qui calcule la factorielle d'un nombre donné.


```python
number = 5
result = 1
for i in list(range(1,(number + 1))):
   result = result * i
print(result)
```
