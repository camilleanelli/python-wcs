```python
borne1 = input("Borne 1: ")
borne2 = input("Borne 2: ")
enter = input("Entrée: ")
values = [borne1, borne2]
min_value = min(values)
max_value = max(values)

while True:
    if (enter > min_value and enter < max_value):
        print(f"Tu as entré {enter}, qui est dans l'intervalle")
        break
    elif enter < min_value:
        print(f"Tu as entré {enter}, plus petit que {min_value}, la borne inférieure")
    elif enter > max_value:
        print(f"Tu as entré {enter}, plus grand que {max_value}, la borne supérieure")
    else:
        print('invalide')
    enter = input("Entrée: ")

```
