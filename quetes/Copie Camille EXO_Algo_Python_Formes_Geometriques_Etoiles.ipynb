# Exercices: Formes Géométriques avec des Étoiles
Ce notebook contient des exercices pour dessiner des formes géométriques avec des étoiles (*) en Python.

## Exercice 1: Triangle en étoiles
**Pseudo-code:**
- Lire un nombre `n`
- Pour chaque ligne `i` allant de 1 à `n`, afficher `i` étoiles

**Instruction:**
Écris un programme qui dessine un triangle avec `n` lignes d'étoiles.

```python
n = int(input("Entrez le nombre de lignes pour le triangle: "))
for i in range(1, n + 1):
  print("*" * i)

```

## Exercice 2: Carré en étoiles
**Pseudo-code:**
- Lire un nombre `n`
- Afficher un carré de taille `n x n` avec des étoiles

**Instruction:**
Écris un programme qui dessine un carré de `n` étoiles de côté.

```python
n = int(input("Entrez la taille du carré: "))
size = 1
while size <= n:
  print("*" * n)
  size += 1
```

## Exercice 3: Pyramide en étoiles
**Pseudo-code:**
- Lire un nombre `n`
- Pour chaque ligne `i` allant de 1 à `n`, afficher des espaces suivis d'étoiles pour former une pyramide

**Instruction:**
Écris un programme qui dessine une pyramide avec `n` lignes d'étoiles.

```python
n = int(input("Entrez le nombre de lignes de la pyramide: "))
stars_counter = 1
i = 1
while i <= n:
  spaces = (((2 * n) - 1 - stars_counter)//2) * " "
  print(spaces, "*" * stars_counter, spaces)
  stars_counter += 2
  i += 1
```

## Exercice 4: Losange en étoiles
**Pseudo-code:**
- Lire un nombre `n`
- Afficher une forme de losange avec des étoiles

**Instruction:**
Écris un programme qui dessine un losange en étoiles de taille `n`.

```python
n = int(input("Entrez la taille du losange (nombre impair): "))
stars = 1
while stars <= n:
  spaces = ((n - stars) // 2) * " "
  print(spaces, "*" * stars, spaces)
  stars += 2

stars2 = n - 2
while stars2 >= 1:
  spaces = ((n - stars2) // 2) * " "
  print(spaces, "*" * stars2, spaces)
  stars2 -= 2
```
