# Partie Bonus

Créez une fonction `analyser_texte(texte)` qui prend une chaîne de caractères en entrée et retourne un dictionnaire contenant les informations suivantes :

- Nombre total de mots
- Nombre de caractères (espaces inclus)
- Nombre de caractères (espaces exclus)
- Mot le plus long
- Longueur moyenne des mots

Testez cette fonction avec le texte suivant :
"Python est un langage de programmation puissant et facile à apprendre. Il dispose de structures de données de haut niveau efficaces et d'une approche simple mais efficace de la programmation orientée objet."

```python
# count the number of words
def nbr_words(text):
  return len(text.split())

# count the number of char with spaces
def nbr_char_w_spaces(text):
  return len(text)

# count the number of char without spaces
def nbr_char_wo_space(text):
  return len(text.replace(' ', ''))

# find the longest word
def max_word(text):
  list_words = text.split()
  dict_words = {}
  for w in list_words:
    dict_words[len(w)] = w
  max_v = max(list(dict_words.keys()))
  return dict_words[max_v]

# calculate the average number char inside words
def avg_word_lenght(txt):
  list_of_len = []
  list_words = txt.split()
  for w in list_words:
    list_of_len.append(len(w))
  return sum(list_of_len) / len(list_words)

def analyser_texte(text):
  resultat = {}
  resultat["Nombre total de mots"] = nbr_words(text)
  resultat["Nombre de caractères (espaces inclus)"] = nbr_char_w_spaces(text)
  resultat["Nombre de caractères (espaces exclus)"] = nbr_char_wo_space(text)
  resultat["Mot le plus long"] = max_word(text)
  resultat["Longueur moyenne des mots"] = avg_word_lenght(text)
  return resultat

print(analyser_texte("Python est un langage de programmation puissant et facile à apprendre. Il dispose de structures de données de haut niveau efficaces et d'une approche simple mais efficace de la programmation orientée objet."))
```
